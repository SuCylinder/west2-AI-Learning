# Logistic函数图像研究
## (1)标准Logistic函数图像

```python
import math
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-100,100,0.5)

def func(L,k,x0,x):
    return L/(1+np.exp(-k*(x-x0)))

y = func(1,1,0,x)

plt.plot(x,y)
plt.show()
```

## (2)不同L,k,&x_0&的Logistic函数图像

### 1.L,k的正负对函数图像的影响

```python
fig ,axs = plt.subplots(3,3)
fig.set_size_inches(10,10)

for i in range(0,len(axs)):
    for j in range(0,len(axs[i])):
        y = func((i-1),(j-1),0,x)
        axs[i][j].plot(x,y)
        axs[i][j].set_title(f"L={i-1},k={j-1}")
```

- $L*k$的正负影响图形是s型还是z型
  - $L*k > 0$图形为s型
  - $L*k < 0$图形为z型
  - $L*k = 0$图形为一条水平线
- $L$影响图形的值域

### 2.L,k的取值的大小对函数图像的影响

```python
fig ,axs = plt.subplots(3,3)
fig.set_size_inches(10,10)

for i in range(0,len(axs)):
    for j in range(0,len(axs[i])):
        y = func(math.pow(10,-i),math.pow(10,-j),0,x)
        axs[i][j].plot(x,y)
        axs[i][j].set_title(f"L={math.pow(10,-i)},k={math.pow(10,-j)}")
```

可以看出
- L影响函数的值域,函数的值域为[0,L]
- k影响图形弯曲程度
  - k越小,图形弯曲程度越小,接近一条直线
  - k越大,图形弯曲程度越大,k很大时,图形几乎在$x=0$处直接从最小值突变到最大值

### 3.$x_0$的取值对函数图像的影响

```python
fig ,axs = plt.subplots(3)
fig.set_size_inches(10,9)

for i in range(0,len(axs)):
        y = func(1,1,(i-1)*50,x)
        axs[i].plot(x,y)
        axs[i].set_title(f"x0 = {(i-1)*50}")
```

可以看出,$x_0$影响函数的中心轴,函数的中心就为$x_0$
