1. 数据生成

```python
import numpy as np

# 生成数据

point_A = np.random.randint(0,100,10).reshape(5,2)
point_B = np.random.randint(0,100,16).reshape(8,2)
print(point_A)
print(point_B)
```

2. 距离矩阵计算

```python
# 扩展维度触发广播
point_A = point_A.reshape(5,1,2)
point_B = point_B.reshape(1,8,2)

d = point_A - point_B

distance_matrix = np.sqrt(np.sum(d**2,2))

print(distance_matrix)
```

3. 轴向数据聚合

```python
min_distance = np.min(distance_matrix,1)
print(min_distance)
```

4. 复合条件查询

```python
x = distance_matrix < 20
print(np.where(x)[1])
```
