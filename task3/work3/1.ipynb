```python
import numpy as np

data_matrix = np.arange(0,100).reshape(10,10)
print(data_matrix)
```

```python
child_matrix = data_matrix[3:7,3:7]
print(child_matrix)
```

```python
data_matrix[data_matrix>75] = 0
print(data_matrix)
```

```python
data_matrix = data_matrix*0.8
print(data_matrix)
```

```python
idx = np.argmax(data_matrix)
idx = np.unravel_index(idx,(10,10))
print("行和列索引")
print(idx[0],idx[1])
print("最大值")
print(data_matrix[idx])
```
