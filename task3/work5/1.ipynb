```python
import numpy as np

grayscale_image = np.random.randint(0,256,size=(200,300))
print(grayscale_image)
```

```python
grayscale_image = grayscale_image.reshape(200,300)
color_image = np.stack([grayscale_image,grayscale_image,grayscale_image],axis=2)
print(color_image)
```

```python
color_image = color_image.reshape(200,300,3,1)

sepia_matrix = np.array([
    [0.393, 0.769, 0.189],
    [0.349, 0.686, 0.168],
    [0.272, 0.534, 0.131]
])

color_image = sepia_matrix @ color_image
# 变换为原来的形状并截断
color_image = np.clip(color_image.reshape(200,300,3),0,255)
print(color_image)
```

```python
R = color_image[...,0]
G = color_image[...,1]
B = color_image[...,2]
# 计算亮度
Luminance = 0.299*R + 0.587*G + 0.114*B
#扩展亮度维度,为广播准备
Luminance = Luminance.reshape(200,300,1)
#应用饱和度增强公式
color_image = Luminance +1.5*(color_image - Luminance)
color_image = np.clip(color_image,0,255)
print(color_image)
```

```python
# 生成左右边框
left = color_image[:,0,:]
right = color_image[:,-1,:]

left = np.linspace(0,left,axis=1,num=20)
right = np.linspace(right,255,axis=1,num=20)

# 拼接
color_image = np.concatenate([left,color_image,right],axis=1)

print(color_image)
```
